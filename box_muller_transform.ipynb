```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
u1 = np.random.uniform(0,1,10000)
sns.distplot(u1)
```

```python
u2 = np.random.uniform(0, 1, 10000)
sns.distplot(u2)
```

```python
theta = 2*math.pi*u1
```

```python
r = np.sqrt(-2*np.log(u2))
```

```python
g1 = r*np.cos(theta)
g2 = r*np.sin(theta)
```

```python
%matplotlib inline
```

```python
sns.distplot(g1)
```

```python
sns.distplot(g2)
```
